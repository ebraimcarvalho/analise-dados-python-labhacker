## Análise dos dados de reembolso dos senadores - 2018

```python
# Importando Bibliotecas
import pandas as pd
```

```python
df = pd.read_csv('2018.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')

df.head()
```

```python
# Verificando tamanho e tipo de dados
df.info()
```

```python
# Análise quantitativa dos dados do dataframe
df.describe()
```

```python
# Os maiores valores reembolsado

df.nlargest(5, 'VALOR_REEMBOLSADO')
```

```python
# Transpões os eixos
df.nlargest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Os 5 menores reembolsos
df.nsmallest(15, 'VALOR_REEMBOLSADO')
```

```python
# Qual o valor total de reembolso
df['VALOR_REEMBOLSADO'].sum().astype(int)
```

```python
# Quantos reembolsos cada senador solicitou?
df['SENADOR'].value_counts()
```

```python
df_agg = df.groupby(['SENADOR']).agg({'VALOR_REEMBOLSADO':sum})
```

```python
g = df_agg['VALOR_REEMBOLSADO'].groupby(level=0, group_keys=False)
```

```python
g.nlargest(30)
```

```python
# Quanto cada Senador solicitou de reembolso em 2018?
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)[0:30]
```

```python
# Quais os tipos de despesa mais reembolsados?
df['TIPO_DESPESA'].value_counts()
```

```python
# E por tipo de despesa, quais os maiores em valores monetários?
df.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```
