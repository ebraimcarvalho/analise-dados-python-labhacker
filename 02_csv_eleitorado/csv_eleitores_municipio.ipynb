## Análise do Eleitorado Brasileiro por municipio

```python
# Importando Bibliotecas
import pandas as pd

# Criando Dataframe
df = pd.read_csv('eleitorado_municipio_2020.csv', sep=';', encoding='latin1')

# Mostra o início do dataframe
df.head()
```

```python
# Mostra o início com os eixos transpostos (Quando tem muitas colunas, é uma forma interessante)
df.head().T
```

```python
# Tipos e quantidades dos dados
df.info()
```

```python
# Cidades com mais eleitores identificados com alguma atipicidade
df.nlargest(5, 'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Quantidade de Eleitoras (Feminino): ', df['QTD_ELEITORES_FEMININO'].sum())
print('Quantidade de Eleitores (Masculino): ', df['QTD_ELEITORES_MASCULINO'].sum())
print('Quantidade de Eleitores (Não informado): ', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Cria variáveis com os totais para calcular o percentual
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print('Quantidade de Eleitoras (Feminino): ', ((tot_fem / tot_eleitores) * 100).round(2),'%')
print('Quantidade de Eleitores (Masculino): ', ((tot_masc / tot_eleitores) * 100).round(2),'%')
print('Quantidade de Eleitores (Não informado): ', ((tot_nao / tot_eleitores) * 100).round(2),'%')
```

```python
# Quantos municípios com mais homens que mulheres?
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Cria uma nova coluna para indicar a relação Feminino / Masculino
df['RELACAO_FM'] = (df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO'])
```

```python
df.head().T
```

```python
# Qual os municípios com maior relação Feminino/Masculino
df.nlargest(5, 'RELACAO_FM').T
```

```python
# Quais os municípios com menor relação Fem/Masc
df.nsmallest(5, 'RELACAO_FM').T
```

```python
# Se precisássemos separar o df para apenas dados de municípios do Brasil
# Usamos copy para criar uma copia e nao ficar atrelado ao dataframe original
df_brasil = df[df['CD_PAIS'] == 1].copy()
df_brasil.head()
```

```python
# Vimos que no df original só tinhamos dados de municipios brasileiros
df_brasil.info()
```

```python
# Quais os municípios com menor relação Fem/Masc do novo dataframe
df_brasil.nsmallest(5, 'RELACAO_FM').T
```

```python
# Quais os municípios com maior relação Fem/Masc do novo dataframe
df_brasil.nlargest(5, 'RELACAO_FM').T
```
