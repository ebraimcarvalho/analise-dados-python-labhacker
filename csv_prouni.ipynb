```python
# Importa Biblioteca e Matplotlib inline
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('cursos-prouni.csv')

df.head()
```

```python
# Verificando o tamanho do Dataframe (Linhas, Colunas)
df.shape
```

```python
# Verifica os tipos de dados
df.dtypes
```

```python
# Analise matematica das colunas numericas
df.describe()
```

```python
# Descreve os dados
df.info()
```

```python
# Quantidade de cursos por UF
df['uf_busca'].value_counts()
```

```python
# Qual o curso com valor de mensalidade mais cara?
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# Qual o curso com valor de mensalidade mais barata e mostra apenas algumas colunas
df[df['mensalidade'] == df['mensalidade'].min()][['grau', 'mensalidade', 'curso_busca', 'uf_busca', 'universidade_nome']]
```

```python
# Qual mensalidade mais cara por UF?
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Qual mensalidade mais barata por UF?
df.groupby('uf_busca')['mensalidade'].min()
```

```python
# Mostrar todos os dados dos cursos mais caros por UF
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```python
# Mostrar todos os dados dos cursos mais baratos por UF e seleciona algumas colunas
df.loc[df.groupby('uf_busca')['mensalidade'].idxmin()][['grau', 'mensalidade', 'curso_busca', 'uf_busca', 'universidade_nome']]
```

```python
# Cursos entre R$500,00 e R$1.000,00
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)]
```

```python
# Retorna as 5 maiores mensalidades
df.nlargest(5, 'mensalidade')
```

```python
# Retorna as 5 menores mensalidades
df.nsmallest(5, 'mensalidade')
```
