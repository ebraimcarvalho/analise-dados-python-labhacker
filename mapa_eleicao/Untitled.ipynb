```python
# Carrega csv e mostra as 5 primeiras linhas do dataaframe
df_votacao = pd.read_csv('dados_eleicao/votacao_candidato_munzona_2018/votacao_candidato_munzona_2018_BR.csv',
                        sep=';', encoding='latin1')

df_votacao.head()
```

```python
df_votacao.info()
```

```python
# Separar os dados para um Dataframe com as informações apenas do 2ª Turno
df_2turno = df_votacao[df_votacao['NR_TURNO'] == 2]

df_2turno.sample(3).T
```

```python
df_2turno.info()
```

```python
# Verificar quantos municipios unicos existem no dataframe
df_2turno['CD_MUNICIPIO'].nunique()
```

```python
# Mas o Brasil tem 5.570 municípios, então precisaremos retirar as cidades do exterior
# vemos que tem 171 municipios fora do brasil

df_2turno[df_2turno['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
# Dessa forma, teremos 5.570 municipios de fato
5741 - 171
```

```python
# Então retiramos os dados do municipio com sigla ZZ e vemos que agora temos 5570 municipios na base
df_2turno = df_2turno[df_2turno['SG_UF'] != 'ZZ']
df_2turno['CD_MUNICIPIO'].nunique()
```

```python
# quero saber quantas zonas tem no municipio de Belo Horizonte
# Vemos que em Belo Horizonte tem 18 zonas eleitorais
df_2turno[df_2turno['NM_MUNICIPIO'] == 'BELO HORIZONTE']['NR_ZONA'].nunique()
```
