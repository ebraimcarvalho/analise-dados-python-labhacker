```python
# Vamos eliminar algumas colunas do dataframe que nao serao utilizadas
cols = ['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']
df_2turno = df_2turno[cols]

# Mostrando os 10 maiores registros de votos nominais
df_2turno.nlargest(10, 'QT_VOTOS_NOMINAIS')
```

```python
# Vamos agora ordenar o dataframe pela quantidade de votos nominais e preservar apenas o primeiro registro
# pois esse será o candidato que ganhou em tal municipio

df_2turno.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)
```

```python
# Vamos ordenar de fato e salvar no lugar
df_2turno.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

# Agora dropamos as linhas com codigo de municipio duplicado, mantendo apenas a primeira linha
df_2turno.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

# Mostra os 10 primeiros
df_2turno.head(10)
```

```python
# Vamos verificar se Jõao pessoa tem apenas uma linha com o candidato vencedor
df_2turno[df_2turno['NM_MUNICIPIO'] == 'JOÃO PESSOA']
```

```python
# Criar uma nova coluna com True se Bolsonaro ou False se Haddad, para facilitar as cores no mapa
df_2turno['RESULTADO'] = df_2turno['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
# Os códigos são diferentes, e se tentarmos encontrar pelo nome do município daria certo?
# Não porque existem municípios com o mesmo nome em cidades diferentes

df_2turno[df_2turno['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```python
# Recuperar a tabela de correspondencia entre IBGE e TSE
# direto do github do Estadão
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia = pd.read_csv(link)

df_equivalencia.head()
```

```python
df_2turno[df_2turno['NM_MUNICIPIO'] == 'ACRELÂNDIA']
```

```python
# Vamos reindexar o dataframe de votação pelo código do TSE
df_vot_novo = df_2turno.copy()

df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

df_vot_novo.head()
```

```python
# Vamos reindexar também o dataframe de equivalencia, pelo codigo do TSE
df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```python
# Verificando o tamanho dos dataframes
df_vot_novo.info()
```

```python
df_equi_novo.info()
```

```python
# Criando um novo dataframe com base no merge de Votação e Equivalencia
df_vot_equi = df_vot_novo.join(df_equi_novo)

df_vot_equi.head()
```

```python
# Vamos juntar o dataframe do GeoPandas com o dataframe Votação/Equivalência

df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# Transformar a coluna de codigo de IBGE do dataframe de Votação/Equivalência para string,
# pois esse valor é string no dataframe do geopandas

df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```python
df_mapa.head()
```

```python
# Vamos unificar os indices

# Mudando o indice do dataframe de Mapa para o codigo do ibge do município
df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)

df_mapa.head()
```

```python
df_vot_equi.head()
```

```python
# Mudando o indice do dataframe de Votação/Equivalência para o codigo do IBGE, assim como do mapa
df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)

df_vot_equi.head()
```

```python
# Vamos juntar os dataframes
df_mapa_novo = df_mapa.join(df_vot_equi)

df_mapa_novo.head().T
```

```python
# Verificando o tamanho do mapa novo
df_mapa_novo.shape
```

```python
df_mapa_novo[df_mapa_novo['NM_MUNICIPIO'].isnull()].head()
```

```python
# verifiando o tamanho do dataframe se selecionarmos as linhas que nao tenha dados null na coluna NM_MUNICIPIO
df_mapa_novo[df_mapa_novo['NM_MUNICIPIO'].notnull()].shape
```

```python
# Retirando as duas linhas onde tem registros NaN, que vimos que são o município lagoa mirim e lagoa dos patos
# Na verdade, essas lagoas não são municipios, mas o shapefile os consideram para desenhar o mapa completo.
df_mapa_novo = df_mapa_novo[df_mapa_novo['NM_MUNICIPIO'].notnull()]

# Também poderíamos usar o dropna, mas pra isso, precisamos ter certeza que nao tem registros em alguma coluna com NaN
# df_mapa_novo.dropna(inplace=True) 

df_mapa_novo.shape
```

```python
# Vamos desenhar o mapa

# primeiro criamos referencia para o grupo de cor, 0/false/red, 1/true/green
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red', 'green'])

# Desenhamos o mapa
fig, ax = plt.subplots(1, figsize=(16, 16))

ax = df_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=0.1, ax=ax, edgecolor='grey')

# Escondendo os axis
ax.set_axis_off()

# Colocando um título
ax.set_title('Resultado Eleição 2018 por Município / 2º Turno', fontdict={'fontsize': '25', 'fontweight': '3'})

# Criando uma nota de rodape
ax.annotate('Fonte: IBGE, 2015; TSE, 2019; Jornal o Estado de São Paulo, 2019',
    xy=(0.1, 0.08), xycoords='figure fraction', horizontalalignment='left',
    verticalalignment='top', fontsize=10, color='#555555')

# Alterando a legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
